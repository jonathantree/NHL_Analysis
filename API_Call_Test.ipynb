```python
import requests

# Import the API key.
from config import API_Key
import pandas as pd
from pprint import pprint 
```

## Team Stats by Season

```python
season_year = '2022'
key = API_Key

url = f"https://api.sportsdata.io/v3/nhl/scores/json/TeamSeasonStats/{season_year}?key={key}"
```

```python
team_data_raw = requests.get(url)
```

```python
team_data_json = team_data_raw.json()
```

```python
season = "2022"
url = f"https://api.sportsdata.io/v3/nhl/stats/json/PlayerSeasonStats/{season}?key={key}"
player_season_stats_raw = requests.get(url)
player_season_stats_json = player_season_stats_raw.json()
```
